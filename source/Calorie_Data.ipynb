```python
import pandas as pd

# ✅ 체형별 칼로리 데이터
calorie_dict = {
    "사과형": {"diet": {"남성": 1500, "여성": 1300}, "maintain": {"남성": 1800, "여성": 1600}, "gain": {"남성": 2500, "여성": 2200}},
    "배형": {"diet": {"남성": 1600, "여성": 1400}, "maintain": {"남성": 1900, "여성": 1700}, "gain": {"남성": 2600, "여성": 2300}},
    "모래시계형": {"diet": {"남성": 1700, "여성": 1500}, "maintain": {"남성": 2000, "여성": 1800}, "gain": {"남성": 2700, "여성": 2400}},
    "엉덩이형": {"diet": {"남성": 1800, "여성": 1600}, "maintain": {"남성": 2100, "여성": 1900}, "gain": {"남성": 2800, "여성": 2500}},
    "상체형": {"diet": {"남성": 1900, "여성": 1700}, "maintain": {"남성": 2200, "여성": 2000}, "gain": {"남성": 2900, "여성": 2600}},
    "하체형": {"diet": {"남성": 2000, "여성": 1800}, "maintain": {"남성": 2300, "여성": 2100}, "gain": {"남성": 3000, "여성": 2700}},
    "표준 체형": {"diet": {"남성": 1700, "여성": 1500}, "maintain": {"남성": 2000, "여성": 1800}, "gain": {"남성": 2700, "여성": 2400}},
}

# ✅ 탄단지 비율 설정
macro_ratio = {
    "diet": {"탄수화물": 40, "단백질": 40, "지방": 20},
    "maintain": {"탄수화물": 50, "단백질": 30, "지방": 20},
    "gain": {"탄수화물": 55, "단백질": 35, "지방": 10},
}

# ✅ 데이터 저장 리스트
data = []

# ✅ 데이터셋 생성
for body_type, goals in calorie_dict.items():
    for goal, kcal_values in goals.items():
        for gender, kcal in kcal_values.items():
            data.append([
                0 if gender == "남성" else 1,  # 성별 (0: 남성, 1: 여성)
                body_type,  # 체형
                goal,  # 목표 (diet / maintain / gain)
                kcal,  # 총 섭취 칼로리
                macro_ratio[goal]["탄수화물"],  # 탄수화물 비율
                macro_ratio[goal]["단백질"],  # 단백질 비율
                macro_ratio[goal]["지방"],  # 지방 비율
            ])

# ✅ DataFrame 생성
df = pd.DataFrame(data, columns=["성별", "체형", "목표", "총 섭취 칼로리", "탄수화물(%)", "단백질(%)", "지방(%)"])

# ✅ CSV 파일 저장
df.to_csv("diet_dataset.csv", index=False, encoding="utf-8-sig")

print("✅ 데이터셋 저장 완료! (diet_dataset.csv)")
```

```python
def convert_serving_size(value):
    match = re.search(r"(\d+)", str(value))  # 숫자만 추출
    return int(match.group(1)) if match else None  # 변환 실패 시 None 반환

food_df["1회 섭취량 (g)"] = food_df["1회 섭취량"].apply(convert_serving_size)

# ✅ 숫자로 변환 (칼로리, 탄수화물, 단백질, 지방)
numeric_cols = ["칼로리", "탄수화물", "단백질", "지방"]
food_df[numeric_cols] = food_df[numeric_cols].apply(pd.to_numeric, errors="coerce")  # 변환 불가 값 NaN 처리

# ✅ NaN 값 확인 및 제거
print("🚀 결측값 개수:\n", df.isna().sum())  
food_df = food_df.dropna()  # NaN 값 제거

# ✅ 전처리된 데이터 확인
print(food_df.head())

# ✅ 전처리된 데이터 저장
food_df.to_csv("data/processed_food_nutrition_data.csv", index=False, encoding="utf-8-sig")
print("✅ 전처리 완료! (processed_food_nutrition_data.csv 저장)")
```

```python
# ✅ 탄단지 비율 계산 함수
def calculate_macros(df):
    df["탄수화물 비율 (%)"] = (df["탄수화물"] * 4) / df["칼로리"] * 100
    df["단백질 비율 (%)"] = (df["단백질"] * 4) / df["칼로리"] * 100
    df["지방 비율 (%)"] = (df["지방"] * 9) / df["칼로리"] * 100
    return df

# ✅ 데이터 로드
file_path = "data/processed_food_nutrition_data.csv"
df = pd.read_csv(file_path, encoding="utf-8-sig")

# ✅ 탄단지 비율 계산 적용
df = calculate_macros(df)

# ✅ 탄단지 비율을 소수점 1자리로 반올림
df["탄수화물 비율 (%)"] = df["탄수화물 비율 (%)"].round(1)
df["단백질 비율 (%)"] = df["단백질 비율 (%)"].round(1)
df["지방 비율 (%)"] = df["지방 비율 (%)"].round(1)

# ✅ 저장
processed_file_path = "data/processed_food_nutrition_data_with_macros_rounded.csv"
df.to_csv(processed_file_path, index=False, encoding="utf-8-sig")

# ✅ 결과 확인
print(df[["식품명(한글)", "칼로리", "탄수화물 비율 (%)", "단백질 비율 (%)", "지방 비율 (%)"]].head())
print(f"✅ 전처리 완료! 새로운 데이터 저장: {processed_file_path}")
```

```python
# 음식 데이터 불러오기
file_path = "data/food_nutrition_data.csv"
df = pd.read_csv(file_path)

# 탄단지 비율 계산 (총합으로 나눠서 비율 구하기)
df["총합"] = df["탄수화물"] + df["단백질"] + df["지방"]
df["탄수화물 비율"] = (df["탄수화물"] / df["총합"]) * 100
df["단백질 비율"] = (df["단백질"] / df["총합"]) * 100
df["지방 비율"] = (df["지방"] / df["총합"]) * 100

# 카테고리 분류
balanced_meals = df[(df["탄수화물 비율"] >= 30) & (df["탄수화물 비율"] <= 40) &
                    (df["단백질 비율"] >= 30) & (df["단백질 비율"] <= 40) &
                    (df["지방 비율"] >= 20) & (df["지방 비율"] <= 40)]

high_protein_meals = df[df["단백질 비율"] > 40]

low_carb_meals = df[df["탄수화물 비율"] < 30]

# 결과 저장
balanced_meals.to_csv("data/balanced_meals.csv", index=False, encoding="utf-8-sig")
high_protein_meals.to_csv("data/high_protein_meals.csv", index=False, encoding="utf-8-sig")
low_carb_meals.to_csv("data/low_carb_meals.csv", index=False, encoding="utf-8-sig")

print("✅ 음식 데이터 분류 완료! (아침, 점심, 저녁 식단용)")
```

```python
calorie_dict = {
    "사과형": {
        "diet": {"남성": 1500, "여성": 1300},
        "maintain": {"남성": 1800, "여성": 1600},
        "muscle_gain": {"남성": 2500, "여성": 2200},
    },
    "배형": {
        "diet": {"남성": 1600, "여성": 1400},
        "maintain": {"남성": 1900, "여성": 1700},
        "muscle_gain": {"남성": 2600, "여성": 2300},
    },
    "모래시계형": {
        "diet": {"남성": 1700, "여성": 1500},
        "maintain": {"남성": 2000, "여성": 1800},
        "muscle_gain": {"남성": 2700, "여성": 2400},
    },
    "엉덩이형": {
        "diet": {"남성": 1800, "여성": 1600},
        "maintain": {"남성": 2100, "여성": 1900},
        "muscle_gain": {"남성": 2800, "여성": 2500},
    },
    "상체형": {
        "diet": {"남성": 1900, "여성": 1700},
        "maintain": {"남성": 2200, "여성": 2000},
        "muscle_gain": {"남성": 2900, "여성": 2600},
    },
    "하체형": {
        "diet": {"남성": 2000, "여성": 1800},
        "maintain": {"남성": 2300, "여성": 2100},
        "muscle_gain": {"남성": 3000, "여성": 2700},
    },
    "표준 체형": {
        "diet": {"남성": 1700, "여성": 1500},
        "maintain": {"남성": 2000, "여성": 1800},
        "muscle_gain": {"남성": 2700, "여성": 2400},
    },
}

# 🔹 목표별 탄단지 비율 설정
macro_ratio = {
    "diet": {"carbs": 0.4, "protein": 0.4, "fat": 0.2},  # 다이어트: 탄40 단40 지20
    "maintain": {"carbs": 0.5, "protein": 0.3, "fat": 0.2},  # 유지: 탄50 단30 지20
    "muscle_gain": {"carbs": 0.5, "protein": 0.4, "fat": 0.1},  # 근육 증가: 탄50 단40 지10
}

def calculate_nutrition(sex, body_type, goal):
    """
    성별, 체형, 목표(다이어트/유지/근육 증가)를 입력받아
    하루 총 섭취 칼로리와 탄단지 비율을 계산하는 함수
    """
    # 🔹 1. 하루 총 섭취 칼로리 가져오기
    total_calories = calorie_dict[body_type][goal][sex]

    # 🔹 2. 목표에 맞는 탄단지 비율 가져오기
    carb_ratio = macro_ratio[goal]["carbs"]
    protein_ratio = macro_ratio[goal]["protein"]
    fat_ratio = macro_ratio[goal]["fat"]

    # 🔹 3. 탄수화물(g), 단백질(g), 지방(g) 계산
    carbs_g = (total_calories * carb_ratio) / 4  # 1g 탄수화물 = 4kcal
    protein_g = (total_calories * protein_ratio) / 4  # 1g 단백질 = 4kcal
    fat_g = (total_calories * fat_ratio) / 9  # 1g 지방 = 9kcal

    # 🔹 4. 결과 반환 (소수점 1자리까지 반올림)
    return {
        "총 섭취 칼로리": total_calories,
        "탄수화물(g)": round(carbs_g, 1),
        "단백질(g)": round(protein_g, 1),
        "지방(g)": round(fat_g, 1),
    }

# ✅ 테스트 실행
sex = "남성"
body_type = "사과형"
goal = "maintain"  # 다이어트 / 유지(maintain) / 근육 증가(muscle_gain)

result = calculate_nutrition(sex, body_type, goal)
print(result)
```
