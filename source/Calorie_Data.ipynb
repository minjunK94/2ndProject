```python
# âœ… íƒ„ë‹¨ì§€ ë¹„ìœ¨ ê³„ì‚° í•¨ìˆ˜
def calculate_macros(df):
    df["íƒ„ìˆ˜í™”ë¬¼ ë¹„ìœ¨ (%)"] = (df["íƒ„ìˆ˜í™”ë¬¼"] * 4) / df["ì¹¼ë¡œë¦¬"] * 100
    df["ë‹¨ë°±ì§ˆ ë¹„ìœ¨ (%)"] = (df["ë‹¨ë°±ì§ˆ"] * 4) / df["ì¹¼ë¡œë¦¬"] * 100
    df["ì§€ë°© ë¹„ìœ¨ (%)"] = (df["ì§€ë°©"] * 9) / df["ì¹¼ë¡œë¦¬"] * 100
    return df

# âœ… ë°ì´í„° ë¡œë“œ
file_path = "data/processed_food_nutrition_data.csv"
df = pd.read_csv(file_path, encoding="utf-8-sig")

# âœ… íƒ„ë‹¨ì§€ ë¹„ìœ¨ ê³„ì‚° ì ìš©
df = calculate_macros(df)

# âœ… íƒ„ë‹¨ì§€ ë¹„ìœ¨ì„ ì†Œìˆ˜ì  1ìë¦¬ë¡œ ë°˜ì˜¬ë¦¼
df["íƒ„ìˆ˜í™”ë¬¼ ë¹„ìœ¨ (%)"] = df["íƒ„ìˆ˜í™”ë¬¼ ë¹„ìœ¨ (%)"].round(1)
df["ë‹¨ë°±ì§ˆ ë¹„ìœ¨ (%)"] = df["ë‹¨ë°±ì§ˆ ë¹„ìœ¨ (%)"].round(1)
df["ì§€ë°© ë¹„ìœ¨ (%)"] = df["ì§€ë°© ë¹„ìœ¨ (%)"].round(1)

# âœ… ì €ì¥
processed_file_path = "data/processed_food_nutrition_data_with_macros_rounded.csv"
df.to_csv(processed_file_path, index=False, encoding="utf-8-sig")

# âœ… ê²°ê³¼ í™•ì¸
print(df[["ì‹í’ˆëª…(í•œê¸€)", "ì¹¼ë¡œë¦¬", "íƒ„ìˆ˜í™”ë¬¼ ë¹„ìœ¨ (%)", "ë‹¨ë°±ì§ˆ ë¹„ìœ¨ (%)", "ì§€ë°© ë¹„ìœ¨ (%)"]].head())
print(f"âœ… ì „ì²˜ë¦¬ ì™„ë£Œ! ìƒˆë¡œìš´ ë°ì´í„° ì €ì¥: {processed_file_path}")
```
